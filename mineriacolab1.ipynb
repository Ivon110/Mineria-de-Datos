####Conceptos

La minería de datos es el proceso de extraer información útil a partir de una acumulación de datos, a menudo provenientes de un almacén o de una recopilación de data sets vinculados. Las herramientas para minería de datos incluyen potentes capacidades estadísticas, matemáticas y analíticas cuyo propósito principal es examinar grandes data sets para identificar tendencias, patrones y relaciones a fin de dar soporte a la toma de decisiones y planificación fundamentadas.

The Basics of NumPy Arrays

```python
#Importamos NumPy
import numpy as np
```

```python
#Generamos una semilla aleatoria
np.random.seed(0)
# Generamos un arreglo aleatorio de enteros, de tamaño seis
x1 = np.random.randint(10, size=6) #El superior no se toma en cuenta es decir va del 0 al 9
# Generamos un arreglo aleatorio de enteros, de tamaño tres renglones y cuatro columnas
x2 = np.random.randint(10, size=(3,4))
# Para hacer un arreglo tridimensional
x3 = np.random.randint(10, size=(3,4,5))
```

```python
print("x3 ndim: ", x3.ndim) # ndim nos dice el numero de dimensiones (x3 tiene 3 dimensiones)
print("x3 shape:", x3.shape) # shape imprime su forma, regresa lo que contiene la lista
print("x3 size: ", x3.size) # size es la multiplicacion del shape 3x4=12 x 5 = 60
```

```python
print(x2.dtype) # dtype nos dice el tamaño en bytes de cada arreglo
```

```python
#itemsize: Lista el tamaño (en bytes) de cada arreglo
print("itemsize:", x3.itemsize, "bytes")
# nbytes: Lista el tamaño total (en bytes) del arreglo
print("nbytes:", x3.nbytes, "bytes")
```

```python
# Son digitos entre el 0 y 10 porque le dimos este limite desde el principio (x1 = np.random.randint(10, size=6) )

x1
```

```python
#Colocamos el índice del elemento a extraer, en este caso el primer elemento tiene el índice 0
x1[0]
```

```python
#Colocamos el índice del elemento a extraer, en este caso el quinto elemento tiene el índice 4
x1[4]
```

```python
# Si queremos el ultimo elemento de ndarray (Arreglo de n dimensiones) podemos poner -1 de índice
x1[-1]
```

```python
#Array Multidimensional
x2
```

```python
#Colocamos el índice del elemento a extraer, en este caso el primer elemento tiene el índice 0,0 para un indice multidimensional
x2[0,0]
```

```python
# Renglon es la fila
# Renglon,Columna
x2[0,2]
```

```python
x2[0,0] = 12 # Asignar un nuevo valor al renglon 0 columna 0
x2
```
